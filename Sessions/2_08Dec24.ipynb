```python
import matplotlib.pyplot as plt
from PIL import Image
import requests

# Valid image URL
url = "https://www.w3schools.com/w3images/lights.jpg"

# Fetch and Display the image
response = requests.get(url, stream=True)
response.raw.decode_content = True
img = Image.open(response.raw)

plt.imshow(img)
plt.axis('on')
plt.show()
```

List Comprehension

```python
[x for x in range(1, 20)]
```

```python
result = []
for i in range(1, 51):
    if(i%5 == 0):
        result.append(i)

result
```

```python
# List Comprehension for the above code 
[result for result in range(1, 51) if result%5==0]
```

```python
# Normal Code
result = []

for i in range(1, 6):
    if i%2 == 0:
        result.append(i**2)
    else:
        result.append(i**3)

result
```

```python
# List Comprehension for the above code
[i**2 if (i%2 == 0) else i**3 for i in range(1, 6)]
```

```python
a = [12,30,99,60,71]
b = []
for i in a:
    if(i > 70):
        b.append("Distinction")
    else:
        if (i <= 70 and i>40):
            b.append("Pass")
        else:
            b.append("Fail")

print(b)
```

```python
# List Comprehension for the above code
["Distiction" if(i > 70) else "Pass" if(i <= 70 and i>40) else "Fail" for i in [12,30,99,60,71]]
```
