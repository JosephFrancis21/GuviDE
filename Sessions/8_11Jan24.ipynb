```python
avg_age_result = collection.aggregate([
    {"$match": {"name": {"$ne": "Nirmal"}}},  # Exclude records with name "Nirmal"
    {"$group": {"_id": None, "average_age": {"$avg": "$age"}}}  # Calculate the average age
])

for i in avg_age_result:
    print(i)
```
